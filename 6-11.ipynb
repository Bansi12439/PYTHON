```python
m={}
for i in range(2):
    k=input("Enter keys: ")
    v=int(input("enter values: "))
    m.update({k:v})
print(m)
y=max(m.values())
g=min(m.keys())

for i,j in m.items():
    if j==y:
        print(f"max scores {i} by {j}")
    if j==g:
        print(f"min scores {i} by {j}")
z=sum(m.values())/2
print("Avg : ",z)
```
