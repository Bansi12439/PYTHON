# Longest sunsequence string code

```python
x=["pyt","pytho","python","pyth"]
x.sort(reverse=True)
ans=""
for i in range(len(x[-1])):
    if x[0][i]==x[-1][i]:
        ans=ans+x[0][i]
        continue
    else:
        break
        print("not found")
print("found",ans)
    
   
    
```
