```python
class WordPlay:
    def __init__(self,l):
        self.l=l
        print(self.l)
    def words_with_length(self,length):
        m1=[]
        for i in self.l:
            if len(i)==length:
                m1.append(i)
        print(m1)
    def starts_with(self,c):
        b1=[]
        for i in self.l:
            if i[0]==c:
                b1.append(i)
        print(b1)
    def ends_with(self,c2):
        b2=[]
        for i in self.l:
            if i[-1]==c2:
                b2.append(i)
        print(b2)
    def palindromes(self):
        l1=[]
        for i in self.l:
                if i==i[::-1]:
                    l1.append(i)
        print(l1)
        
    def only(self,s1):
        z=set(s1)
        for i in self.l:
            w=set(i)
            if z.intersection(w)==z:
                print(i)
    def avoids(self,s2):
        l2=[]
        z=set(s2)
        for i in self.l:
            w=set(i)
            if z.intersection(w)==set():
                l2.append(i)
        print(l2)
            
        
w=WordPlay(['appple','banana','find','dictionary','set','tuple','list','malayalam','nayan','grind','apricot'])
w.words_with_length(5)
w.starts_with('a')
w.ends_with('d')
w.palindromes()
w.only('bna')
w.avoids('amkd')
```
