```python
class A:
    def __init__(self):
        print("A")
    def base(self):
        print("parent child")
class B(A):
    def __init__(self):
        print("Child")
        super().__init__()
    def show(self):
        print("B")
        super().base()
b=B()
b.show()
```

```python
class A:
    def __init__(self,a,b):
        self.a=a
        self.b=b
        print(self.a,self.b)
        print("A")
    def base(self):
        print("parent child")
class B(A):
    def __init__(self,c,a,b):
        super().__init__(a,b)
        self.c=c
        print(self.a)
        print("Child")
       
    def show(self):
        print("B")
        super().base()
b=B(30,60,50)
b.show()
```

# object operator

```python
class A:
    def __init__(self,a):
        self.a=a
        print(self.a)
    def __mul__(self,other):
        print(self.a*other.a)
    def __str__(self):
        return "hello"
    def __pow__(self,other):
        print(self.a**other.a)
    def __eq__(self,other):
        print(self.a==other.a)
    def __truediv__(self,other):
         print(self.a/other.a)
        
a=A(2)
b=A(3)
print(b)
a*b
a**b
a==b
a/b
```

```python
from abc import ABC,abstractmethod
class A(ABC):
    @abstractmethod
    def demo(self):
        pass
    def show(self):
        print("num")
class B(A):
    def demo(self):
        print("B")
class C(B):
    def show(self):
        print("c")
a=B()
a.demo()
a.show()
c=C()
c.show()
```
